**Ploting**

```
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(house_price_df['price'], bins=50, kde=True)
plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

**Conclusion:**
The plot shows that most houses are sold at lower prices, mostly below ₹1 million. Very few houses have very high prices, which are outliers. The shape of the graph is skewed to the right. This means house prices are not spread equally and some very expensive homes affect the average.

```
features = ['sqft_living', 'bedrooms', 'bathrooms', 'floors', 'sqft_lot', 'grade', 'sqft_basement']

# Calculate rows needed (2 columns per row)
cols = 2
rows = (len(features) + 1) // cols

# Create subplots
fig, axes = plt.subplots(rows, cols, figsize=(12, 5 * rows))
axes = axes.flatten()

# Loop through features and plot
for i, feature in enumerate(features):
    sns.scatterplot(data=house_price_df, x=feature, y='price', ax=axes[i])
    axes[i].set_title(f"Price vs {feature}")

# Hide unused subplots if any
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

**Conclusion:**

1. **Price vs sqft_living:**

  -> As living area increases, the price also increases.

  -> There is a strong positive relationship between square footage of the living area and price.
  
  -> Bigger houses generally cost more.

2. **Price vs bedrooms:**

   -> Houses with 3 to 5 bedrooms are the most common.

   -> More bedrooms don’t always mean higher price.

   -> Extremely high bedroom counts (like 10+) don't always have high prices, indicating possible outliers.

3. **Price vs bathrooms:**
   
   -> Price tends to increase with more bathrooms.

   -> Homes with 2 to 4 bathrooms are common and generally priced higher than homes with 1 bathroom.

   -> The trend is positive, but with some spread.

4. **Price vs floors:**
   -> Most homes have 1 or 2 floors.

  -> Price doesn’t show a strong trend with number of floors.

  -> More floors don't necessarily lead to higher price.

5. **Price vs sqft_lot:**
   -> There's no clear pattern between lot size and price.

   -> Some very large lots don’t have high prices.

   -> A few outliers have huge lot sizes but average prices.

6. **Price vs grade:**

   -> There is a clear upward trend: as the grade (quality) of the house increases, the price also increases.

   -> This is one of the strongest positive relationships observed.

7. **Price vs sqft_basement:**

   -> Homes with larger basements can have higher prices, but the relationship is not very strong.

   -> Many homes have no basement (0 sqft).

```
sns.barplot(x='bedrooms', y='price', data=house_price_df)
plt.title("Average Price by Number of Bedrooms")
plt.show()

sns.barplot(x='bathrooms', y='price', data=house_price_df)
plt.title("Average Price by Number of Bathrooms")
plt.show()
```

**Conclusion:** As the number of bedrooms or bathrooms increases, the average house price also increases. However, after a point, the price becomes less stable due to fewer houses with very high counts.

```
# Group by year directly using dt.year
yearly_prices = house_price_df.groupby(house_price_df['date'].dt.year)['price'].mean().reset_index()

# Plot
sns.lineplot(data=yearly_prices, x='date', y='price')
plt.title("Average Price Per Year")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.show()
```

**Conclusion:** The line plot shows that the average house price increased slightly from 2014 to 2015. This suggests a small upward trend in house prices over the one-year period.

```
sns.boxplot(x='grade', y='price', data=house_price_df)
plt.title("Price Distribution Across House Grades")
plt.show()
```

**Conclusion:** As the house grade increases, the price also increases. Higher-grade houses usually have higher prices, and the range of prices also becomes wider, showing more expensive homes in top grades.

```
top_zipcodes = house_price_df['zipcode'].value_counts().head(10).index
df_top_zip = house_price_df[house_price_df['zipcode'].isin(top_zipcodes)]
sns.boxplot(x='zipcode', y='price', data=df_top_zip)
plt.title("Price Distribution in Top 10 Zip Codes")
plt.xticks(rotation=45)
plt.show()
```

**Conclusion:** The boxplot shows that house prices vary a lot between zip codes. Some areas like 98006 have much higher prices, while others have more affordable and consistent price ranges.

```
plt.figure(figsize=(8, 5))
sns.countplot(x='bedrooms', data=house_price_df)
plt.title("Number of Houses by Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Count")
plt.show()
```

**Conclusion:**
Most houses have 3 or 4 bedrooms. Very few houses have more than 6 or fewer than 2 bedrooms, making 3-bedroom homes the most common.

```
plt.figure(figsize=(10, 6))
sns.scatterplot(data=house_price_df, x='long', y='lat', hue='price', palette='coolwarm', alpha=0.5)
plt.title("House Prices by Location")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(title='Price')
plt.show()
```

**Conclusion:**
House prices change with location. Higher-priced homes are grouped in certain areas, showing that location has a strong effect on housing prices.

```
sns.scatterplot(x='yr_built', y='price', data=house_price_df)
plt.title("Year Built vs Price")
plt.show()
```

**Conclusion:** The plot shows that houses built in recent years tend to have higher prices, but older homes are also sold at various prices. There's no clear trend, but newer homes generally cost more.

```
plt.figure(figsize=(12, 10))
sns.heatmap(house_price_df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

**Conclusion:**
The heatmap shows that price is strongly related to sqft_living, grade, and sqft_above. These features have the highest positive correlation, so they are most useful for predicting house prices.

Importing required libraries.

```
!pip install xgboost

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

Preparing Features and Target

```
x = house_price_df.drop(['price', 'date', 'zipcode', 'lat', 'long'], axis=1)
y = house_price_df['price']
```

Spility the data into train and test sets.

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Performing feature scalling.

```
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

# Linear regression.

```
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
lr_pred = lr.predict(X_test_scaled)
r2_lr = r2_score(y_test, lr_pred)
print("Linear Regression R² Score:", round(r2_lr, 2))
```

# Decision Tree

```
dt = DecisionTreeRegressor(random_state=42)
dt.fit(x_train, y_train)
dt_pred = dt.predict(x_test)
r2_dt = r2_score(y_test, dt_pred)
print("Decision Tree Regression R² Score:", round(r2_dt, 2))
```

# Random Forest

```
rf = RandomForestRegressor(random_state=42)
rf.fit(x_train, y_train)
rf_pred = rf.predict(x_test)
r2_rf = r2_score(y_test, rf_pred)
print("Random Forest Regression R² Score:", round(r2_rf, 2))
```

# Gradient Boosting Regressor

```
gb = GradientBoostingRegressor(random_state=42)
gb.fit(x_train, y_train)
gb_pred = gb.predict(x_test)
r2_gb = r2_score(y_test, gb_pred)
print("Gradient Boosting Regressor R² Score:", round(r2_gb, 2))
```

# XGBoost Regressor

```
xgb = XGBRegressor(random_state=42)
xgb.fit(x_train, y_train)
xgb_pred = xgb.predict(x_test)
r2_xgb = r2_score(y_test, xgb_pred)
print("XGBoost Regressor R² Score:", round(r2_xgb, 2))
```

# **Conclusion**:
In this project, multiple machine learning models were used to predict house prices. Random Forest and Gradient Boosting performed the best, both achieving an R² score of around 0.71. This means they were able to explain about 71% of the variation in house prices. Linear Regression did moderately well, while Decision Tree had the lowest score. Based on these results, ensemble models like **Random Forest and Gradient Boosting** are most suitable for this dataset.
